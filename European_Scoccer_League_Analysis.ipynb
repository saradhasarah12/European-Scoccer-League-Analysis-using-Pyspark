```python
evtTypeMap = {0:'Announcement', 1:'Attempt', 2:'Corner', 3:'Foul', 4:'Yellow card', 5:'Second yellow card', 6:'Red card', 7:'Substitution', 8:'Free kick won', 9:'Offside', 10:'Hand ball', 11:'Penalty conceded'}

evtTyp2Map = {12:'Key Pass', 13:'Failed through ball', 14:'Sending off', 15:'Own goal'}

sideMap = {1:'Home', 2:'Away'}

shotPlaceMap = {1:'Bit too high', 2:'Blocked', 3:'Bottom left corner', 4:'Bottom right corner', 5:'Centre of the goal', 6:'High and wide', 7:'Hits the bar', 8:'Misses to the left', 9:'Misses to the right', 10:'Too high', 11:'Top centre of the goal', 12:'Top left corner', 13:'Top right corner'}

shotOutcomeMap = {1:'On target', 2:'Off target', 3:'Blocked', 4:'Hit the bar'}

locationMap = {1:'Attacking half', 2:'Defensive half', 3:'Centre of the box', 4:'Left wing', 5:'Right wing', 6:'Difficult angle and long range', 7:'Difficult angle on the left', 8:'Difficult angle on the right', 9:'Left side of the box', 10:'Left side of the six yard box', 11:'Right side of the box', 12:'Right side of the six yard box', 13:'Very close range', 14:'Penalty spot', 15:'Outside the box', 16:'Long range', 17:'More than 35 yards', 18:'More than 40 yards', 19:'Not recorded'}

bodyPartMap = {1:'Right foot', 2:'Left foot', 3:'Head'}

assistMethodMap = {0:'None', 1:'Pass', 2:'Cross', 3:'Headed pass', 4:'Through ball'}

situationMap = {1:'Open play', 2:'Set piece', 3:'Corner', 4:'Free kick'}

countryCodeMap = {'germany':'DEU', 'france':'FRA', 'england':'GBR', 'spain':'ESP', 'italy':'ITA'}
```

udf to map key to values in dict

```python
from pyspark.sql.types import StringType
from pyspark.sql.functions import udf

def translateKeys(mapping):
    def translateKeys_(col):
        return mapping.get(col)
    return udf(translateKeys_, StringType())
```

easy access to contry column and spelling mistakes in country names

```python
gameDf = gameDf.withColumn("country_code", translateKeys(countryCodeMap)("country"))

display(gameDf['id_odsp','country','country_code'])
```

```python
eventsDf = (
             eventsDf.
             withColumn("event_type_str", translateKeys(evtTypeMap)("event_type")).
             withColumn("event_type2_str", translateKeys(evtTyp2Map)("event_type2")).
             withColumn("side_str", translateKeys(sideMap)("side")).
             withColumn("shot_place_str", translateKeys(shotPlaceMap)("shot_place")).
             withColumn("shot_outcome_str", translateKeys(shotOutcomeMap)("shot_outcome")).
             withColumn("location_str", translateKeys(locationMap)("location")).
             withColumn("bodypart_str", translateKeys(bodyPartMap)("bodypart")).
             withColumn("assist_method_str", translateKeys(assistMethodMap)("assist_method")).
             withColumn("situation_str", translateKeys(situationMap)("situation"))
           )
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.classification import GBTClassifier
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```python
categFeatures = ["event_type_str", "event_team", "shot_place_str", "location_str", "assist_method_str", "situation_str", "country_code"]

stringIndexers = [StringIndexer().setInputCol(baseFeature).setOutputCol(baseFeature + "_idx") for baseFeature in categFeatures]

encoders = [OneHotEncoder().setInputCol(baseFeature + "_idx").setOutputCol(baseFeature + "_vec") for baseFeature in categFeatures]

featureAssembler = VectorAssembler()
featureAssembler.setInputCols([baseFeature + "_vec" for baseFeature in categFeatures])
featureAssembler.setOutputCol("features")
```

```python
(trainingData, testData) = finalDf.randomSplit([0.75, 0.25])
```

***WITHOUT HYPER PARAMETER TUNING***

*NAIVE BAYES*

```python
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, matthews_corrcoef, cohen_kappa_score

naive_bayes = NaiveBayes(labelCol="is_goal", featuresCol="features", smoothing=1.0)

nb_pipelineStages = stringIndexers + encoders + [featureAssembler, naive_bayes]
nb_pipeline = Pipeline(stages=nb_pipelineStages)

nb_model = nb_pipeline.fit(trainingData)

nb_predictions = nb_model.transform(testData)

nb_true_labels = [int(row.is_goal) for row in nb_predictions.select("is_goal").collect()]
nb_model_predictions = [int(row.prediction) for row in nb_predictions.select("prediction").collect()]

nb_accuracy = accuracy_score(nb_true_labels, nb_model_predictions)
nb_precision = precision_score(nb_true_labels, nb_model_predictions)
nb_recall = recall_score(nb_true_labels, nb_model_predictions)
nb_f1 = f1_score(nb_true_labels, nb_model_predictions)
nb_roc_auc = roc_auc_score(nb_true_labels, nb_model_predictions)

print("Naive Bayes Classifier Metrics:")
print(f"Accuracy: {nb_accuracy:.2f}")
print(f"Precision: {nb_precision:.2f}")
print(f"Recall: {nb_recall:.2f}")
print(f"F1-Score: {nb_f1:.2f}")
print(f"ROC AUC: {nb_roc_auc:.2f}")
```

*SINGULAR VECTOR CLASSIFIER*

```python
from pyspark.ml.classification import LinearSVC

svmClassifier = LinearSVC(labelCol="is_goal", featuresCol="features")

svmPipelineStages = stringIndexers + encoders + [featureAssembler, svmClassifier]

svmPipeline = Pipeline(stages=svmPipelineStages)

svmModel = svmPipeline.fit(trainingData)
svmPredictions = svmModel.transform(testData)
svmEvaluator = BinaryClassificationEvaluator(labelCol="is_goal", rawPredictionCol="prediction")
svmAUC = svmEvaluator.evaluate(svmPredictions)
print("SVM AUC:", svmAUC)
```

*RANDOM FOREST CLASSIFIER*

```python
from pyspark.ml.classification import RandomForestClassifier

rfClassifier = RandomForestClassifier(labelCol="is_goal", featuresCol="features", numTrees=10)

pipelineStages = stringIndexers + encoders + [featureAssembler, rfClassifier]

pipeline = Pipeline(stages=pipelineStages)

(trainingData, testData) = finalDf.randomSplit([0.75, 0.25])
model = pipeline.fit(trainingData)

predictions = model.transform(testData)

display(predictions.select("prediction", "is_goal", "features"))
evaluator = BinaryClassificationEvaluator(labelCol="is_goal", rawPredictionCol="prediction")
evaluation = evaluator.evaluate(predictions)
print("Random Forest AUC:", evaluation)
```

*GRADIENT BOOSTED TREE*

```python
gbtClassifier = GBTClassifier(labelCol="is_goal", featuresCol="features", maxDepth=5, maxIter=20)

pipelineStages = stringIndexers + encoders + [featureAssembler, gbtClassifier]
pipeline = Pipeline(stages=pipelineStages)

(trainingData, testData) = finalDf.randomSplit([0.75, 0.25])
model = pipeline.fit(trainingData)

predictions = model.transform(testData)
display(predictions.select("prediction", "is_goal", "features"))
evaluator = BinaryClassificationEvaluator(
    labelCol="is_goal", rawPredictionCol="prediction")
evaluator.evaluate(predictions)
```

***HYPER PARAMETER TUNING***

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

naive_bayes = NaiveBayes(labelCol="is_goal", featuresCol="features")

nb_pipelineStages = stringIndexers + encoders + [featureAssembler, naive_bayes]
nb_pipeline = Pipeline(stages=nb_pipelineStages)

param_grid = (ParamGridBuilder()
    .addGrid(naive_bayes.smoothing, [0.0, 0.2, 0.4, 0.1])
    .build())

cross_validator = CrossValidator(
    estimator=nb_pipeline,
    estimatorParamMaps=param_grid,
    evaluator=BinaryClassificationEvaluator(labelCol="is_goal", rawPredictionCol="prediction"),
    numFolds=3
)

cv_model = cross_validator.fit(trainingData)

best_nb_model = cv_model.bestModel

best_nb_predictions = best_nb_model.transform(testData)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

best_nb_true_labels = [int(row.is_goal) for row in best_nb_predictions.select("is_goal").collect()]
best_nb_model_predictions = [row.prediction for row in best_nb_predictions.select("prediction").collect()]

best_nb_roc_auc = roc_auc_score(best_nb_true_labels, best_nb_model_predictions)
print(f"ROC AUC: {best_nb_roc_auc:.2f}")
```

```python
from pyspark.ml.classification import RandomForestClassifier

rf_classifier = RandomForestClassifier(labelCol="is_goal", featuresCol="features")

rf_pipelineStages = stringIndexers + encoders + [featureAssembler, rf_classifier]
rf_pipeline = Pipeline(stages=rf_pipelineStages)

param_grid_rf = (ParamGridBuilder()
    .addGrid(rf_classifier.numTrees, [10, 20, 30])
    .addGrid(rf_classifier.maxDepth, [5, 10, 15])
    .build())

cross_validator_rf = CrossValidator(
    estimator=rf_pipeline,
    estimatorParamMaps=param_grid_rf,
    evaluator=BinaryClassificationEvaluator(labelCol="is_goal", rawPredictionCol="prediction"),
    numFolds=3
)

cv_model_rf = cross_validator_rf.fit(trainingData)

best_rf_model = cv_model_rf.bestModel

best_rf_predictions = best_rf_model.transform(testData)

best_rf_true_labels = [int(row.is_goal) for row in best_rf_predictions.select("is_goal").collect()]
best_rf_model_predictions = [row.prediction for row in best_rf_predictions.select("prediction").collect()]

best_rf_roc_auc = roc_auc_score(best_rf_true_labels, best_rf_model_predictions)
print(f"Random Forest ROC AUC: {best_rf_roc_auc:.2f}")
```

```python
from pyspark.ml.classification import LinearSVC

svm_classifier = LinearSVC(labelCol="is_goal", featuresCol="features")

svm_pipelineStages = stringIndexers + encoders + [featureAssembler, svm_classifier]
svm_pipeline = Pipeline(stages=svm_pipelineStages)

param_grid_svm = (ParamGridBuilder()
    .addGrid(svm_classifier.regParam, [0.01, 0.1, 1.0])
    .addGrid(svm_classifier.maxIter, [10, 20, 30])
    .build())

cross_validator_svm = CrossValidator(
    estimator=svm_pipeline,
    estimatorParamMaps=param_grid_svm,
    evaluator=BinaryClassificationEvaluator(labelCol="is_goal", rawPredictionCol="prediction"),
    numFolds=3
)

cv_model_svm = cross_validator_svm.fit(trainingData)

best_svm_model = cv_model_svm.bestModel

best_svm_predictions = best_svm_model.transform(testData)

best_svm_true_labels = [int(row.is_goal) for row in best_svm_predictions.select("is_goal").collect()]
best_svm_model_predictions = [row.prediction for row in best_svm_predictions.select("prediction").collect()]

best_svm_roc_auc = roc_auc_score(best_svm_true_labels, best_svm_model_predictions)
print(f"SVM ROC AUC: {best_svm_roc_auc:.2f}")
```
